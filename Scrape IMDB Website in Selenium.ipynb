```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time as t
import csv
import random
import threading as th
```

```python
# Chrome driver Path
chromepath= r"chromedriver.exe"

option= webdriver.ChromeOptions()      # open google chrome session

option.add_argument('-disk-cache-size= 3000000000')   # OPTIONAL,    --->  make memory cache size = 3000000000                                                                                                                                                                                                                                                                                                                         
driver= webdriver.Chrome(executable_path= chromepath, options= option)   # open driver

# ++++++++++++++++++++++++++++++++++++

driver.get('https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&ref_=adv_prv')

# Scrape Movie Titles:
Titles_lst= []
containers= driver.find_elements_by_css_selector("h3[class= 'lister-item-header']")         
for container in containers:
    titles= container.find_element_by_css_selector("a").text
    Titles_lst.append(titles)

# Scrape Movie Production Years:
Production_Years= []
production_years= driver.find_elements_by_css_selector("span[class= 'lister-item-year text-muted unbold']")           
production_year_lst= [element.text[-5:-1] for element in production_years]
Production_Years.extend(production_year_lst)

# Scrape Movie Certificates:
Certificates= []
certificates= driver.find_elements_by_css_selector("span[class= 'certificate']")            
certificate_lst= [element.text for element in certificates]
Certificates.extend(certificate_lst)

# Scrape Movie Runtime:
Runtimes= []
runtimes= driver.find_elements_by_css_selector("span[class= 'runtime']")            
runtime_lst= [element.text for element in runtimes]
Runtimes.extend(runtime_lst)

# Scrape Movie Genres:
Genres= []
genres= driver.find_elements_by_css_selector("span[class= 'genre']")            
genre_lst= [element.text for element in genres]
Genres.extend(genre_lst)

# Scrape Movie Rates:
Rates_lst= []
containers= driver.find_elements_by_css_selector("div[class= 'inline-block ratings-imdb-rating']")          
for container in containers:
    rates= container.find_element_by_css_selector("strong").text
    Rates_lst.append(rates)

# Scrape Movie Stories:
Stories= []
stories= driver.find_elements_by_css_selector("p[class= 'text-muted']")            
story_lst= [element.text for element in stories]
Stories.extend(story_lst)

# Scrape Movie Director:
Directors= []
for j in range(1, 101):
    directors= driver.find_elements_by_xpath(f'//*[@id="main"]/div/div[3]/div/div[{j}]/div[3]/p[3]/a[1]')            
    director_lst= [element.text for element in directors]
    Directors.extend(director_lst)

# scrape Movie Votes:
votes_lst= []
containers= driver.find_elements_by_css_selector("div[class= 'lister-item-content']")          
for container in containers:
    votes= container.find_element_by_css_selector("p[class= 'sort-num_votes-visible']").find_element_by_css_selector("span:nth-child(2)").text
    votes_lst.append(votes)

# =========================================================================================

for i in range(1, 10):
    driver.get(f'https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start={i}01&ref_=adv_nxt')
      
    # Scrape Movie Titles:
    containers= driver.find_elements_by_css_selector("h3[class= 'lister-item-header']")          
    for container in containers:
        titles= container.find_element_by_css_selector("a").text
        Titles_lst.append(titles)
    
    # Scrape Movie Production Years:
    production_years= driver.find_elements_by_css_selector("span[class= 'lister-item-year text-muted unbold']")            
    production_year_lst= [element.text[-5:-1] for element in production_years]
    Production_Years.extend(production_year_lst)

    # Scrape Movie Certificates:
    certificates= driver.find_elements_by_css_selector("span[class= 'certificate']")            
    certificate_lst= [element.text for element in certificates]
    Certificates.extend(certificate_lst)

    # Scrape Movie Runtime:
    runtimes= driver.find_elements_by_css_selector("span[class= 'runtime']")            
    runtime_lst= [element.text for element in runtimes]
    Runtimes.extend(runtime_lst)

    # Scrape Movie Genres:
    genres= driver.find_elements_by_css_selector("span[class= 'genre']")            
    genre_lst= [element.text for element in genres]
    Genres.extend(genre_lst)

    # Scrape Movie Rates:
    containers= driver.find_elements_by_css_selector("div[class= 'inline-block ratings-imdb-rating']")          
    for container in containers:
        rates= container.find_element_by_css_selector("strong").text
        Rates_lst.append(rates)

    # Scrape Movie Stories:
    stories= driver.find_elements_by_css_selector("p[class= 'text-muted']")            
    story_lst= [element.text for element in stories]
    Stories.extend(story_lst)

    # Scrape Movie Director:
    for j in range(1, 101):
        directors= driver.find_elements_by_xpath(f'//*[@id="main"]/div/div[3]/div/div[{j}]/div[3]/p[3]/a[1]')            
        director_lst= [element.text for element in directors]
        Directors.extend(director_lst)

    # scrape Movie Votes:
    containers= driver.find_elements_by_css_selector("div[class= 'lister-item-content']")          
    for container in containers:
        votes= container.find_element_by_css_selector("p[class= 'sort-num_votes-visible']").find_element_by_css_selector("span:nth-child(2)").text
        votes_lst.append(votes)
        
###########################################################################
        
```

```python
# Save all data in csv files:
from itertools import zip_longest
from csv import writer

file_list= [Titles_lst, Production_Years, Certificates, Runtimes, Genres, Rates_lst, Stories, Directors, votes_lst]
exported= zip_longest(*file_list)

with open(f'IMDB Movies.csv', 'w', encoding="utf-8") as csv_file:
    wr= writer(csv_file)
    wr.writerow(['Titles', 'Production Year', 'Certificates', 'Runtimes', 'Genres', 'Rates', 'Stories', 'Directors', 'Votes'])
    wr.writerows(exported)
```
